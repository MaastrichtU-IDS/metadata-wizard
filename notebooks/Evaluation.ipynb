```python
text_table = pd.read_csv('../datasets/text_in_table.csv')
```

```python
id_list = text_table.pubmed_id_title
title_list = text_table.text_title
```

```python
#keywords_found in titles 
list_records = []
for ids, title in zip(id_list, title_list):
    list_records.append((ids, keyword_processor.extract_keywords(title)))
```

```python
dir_titles = {key: list(set(value)) for key, value in list_records }
```
